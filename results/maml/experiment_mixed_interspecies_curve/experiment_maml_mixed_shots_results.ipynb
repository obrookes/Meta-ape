```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
with open('metacurve_interspecies_results_n5_k1.npy', 'rb') as f:
    meta_train_loss1 = np.load(f)[:50]
    meta_train_acc1 = np.load(f)[:50]
    meta_test_loss1 = np.load(f)[:50]
    meta_test_acc1 = np.load(f)[:50]
```

**Calculating averages**

Average meta-train accuracy across all shots

```python
all_meta_train_acc = np.array([meta_train_acc1, meta_train_acc3, meta_train_acc5])
average_meta_train_acc = all_meta_train_acc.sum(axis=0) / len(all_meta_train_acc)
```

Average meta-test accuracy across all shots

```python
all_meta_test_acc = np.array([meta_test_acc1, meta_test_acc3, meta_test_acc5])
average_meta_test_acc = all_meta_test_acc.sum(axis=0) / len(all_meta_test_acc)
```

```python
# same plotting code as above!
sns.set()
plt.plot(average_meta_train_acc,label='Train')
plt.plot(average_meta_test_acc,label='Test')
plt.title("ANIL: Avg Train Test Accuracy")
plt.xlabel("Iterations")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

```python
def generate_plots(file_path,model,ways,shots):
    with open(file_path, 'rb') as f:
        meta_train_loss = np.load(f)
        meta_train_acc = np.load(f)
        meta_test_loss = np.load(f)
        meta_test_acc = np.load(f)
        
        output_file = model + "-" + "n" + str(ways) + "-" + "k" + str(shots) 
        
        title = model.upper() + " " + str(ways).upper() + "-Way" + " " + str(shots) + "-Shot"
        
        # Plot loss
        sns.set()
        plt.plot(meta_train_loss,label='Train')
        plt.plot(meta_test_loss,label='Test')
        
        loss_title = title + " Loss"
        
        plt.title(loss_title)
        plt.xlabel("Iterations")
        plt.ylabel("Loss")
        plt.legend()
        
        loss_file = output_file + "_" + "loss.pdf"
        
        plt.savefig(loss_file)
        plt.clf()
        
        # Plot accuracy
        plt.plot(meta_train_acc,label='Train')
        plt.plot(meta_test_acc,label='Test')
        
        acc_title = title + " Accuracy"
        
        plt.title(acc_title)
        plt.xlabel("Iterations")
        plt.ylabel("Accuracy")
        plt.legend()
        
        acc_file = output_file + "_" + "acc.pdf"
        
        plt.savefig(acc_file)
```

```python
path = 'anil_intraspecies_results_n5_k3.npy'
generate_plots(path, model='anil', ways=5, shots=3)
```
